```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.isnull()
```

```python
df['species'].unique()
```

```python
df1=df[df['species']!='setosa']
```

```python
df1
```

```python
df1['species']=df1['species'].map({'versicolor':0, 'virginica':1})
```

```python
df1
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
y=df1["species"]
```

```python
X=df1.loc[:,df1.columns!="species"]
```

```python
y.shape
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train
```

```python
y_train
```

logistic regression

```python
log_model=LogisticRegression()
```

```python
log_model.fit(X_train, y_train)
```

```python
y_pred=log_model.predict(X_test)
```

```python
y_pred
```

```python
acc_log=round(accuracy_score(y_pred,y_test)*100,2)
```

```python
print("Accuracy of Logistic Regression",acc_log)
```

##K Nearest Neighbor (KNN)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier()
```

```python
knn_model.fit(X_train, y_train)
```

```python
y_pred=knn_model.predict(X_test)
```

```python
y_pred
```

```python
acc_knn=round(accuracy_score(y_test,y_pred)*100,2)
print("Accuracy of K nearest Neighbors:", acc_knn)
```

##Support Vector Machine ##

```python
from sklearn import svm
```

```python
svm_model=svm.SVC()
```

```python
svm_model.fit(X_train,y_train)
```

```python
y_pred=svm_model.predict(X_test)
```

```python
y_pred
```

```python
acc_svc=round(accuracy_score(y_test,y_pred)*100,2)
print('Accuracy of support vector classifier:',acc_svc)
```

###Decision Tree###

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree_model=DecisionTreeClassifier()
```

```python
dtree_model.fit(X_train,y_train)
```

```python
y_pred=dtree_model.predict(X_test)
```

```python
y_pred
```

```python
acc_dtree=round(accuracy_score(y_test,y_pred)*100,2)
print('Accuracy of Decison Tree Classifier:', acc_dtree)
```

###Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc_model=RandomForestClassifier()
```

```python
rfc_model.fit(X_train,y_train)
```

```python
y_pred=rfc_model.predict(X_test)
```

```python
y_pred
```

```python
acc_rfc=round(accuracy_score(y_test,y_pred)*100,2)
print('Accuracy of Random forest Classifier:', acc_rfc)
```

###Model Selection

```python
models=pd.DataFrame({
    'model':['Logistic Regression','Support vector Machine','K Nearest Neighbor','Decision Tree classifier','Random forest Classifier'],'score':[acc_log, acc_svc, acc_knn, acc_dtree, acc_rfc]
})
```

```python
models.sort_values(by='score',ascending=False)
```
